```python
#what do most clients do
from unidecode import unidecode
import re

def normalize_title(title):
    title = unidecode(title).strip().lower() 
    title = re.sub(r'\s+', '', title)  
    return title
```

```python
client_df['title'] = client_df['title'].apply(normalize_title)
clients_by_title = client_df.groupby('title').size().sort_values(ascending=False)
clients_by_title
```

```python
#city with the most users 
city_mapping = dict(zip(city_df['name'].str.lower(), city_df['_id']))  
```

```python
def map_address_to_city(address):
    if pd.isna(address):
        return "Casablanca"  # Assign Casablanca if address is NaN
    
    address_lower = address.lower()
    
    # Check if the address contains any known city name
    for city_name in city_mapping.keys():
        if city_name in address_lower:
            return city_name.capitalize()
    
    # If no city is found, assign Casablanca by default
    return "Casablanca"
```

```python
#new column 'city' in user_df
user_df['city'] = user_df['address'].apply(map_address_to_city).replace("   ", 'Unknown')
print(user_df[['address', 'city']])
```

```python
#city with the most users 
city_estheticians = user_df.groupby('city').size().sort_values(ascending=False)
city_estheticians
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(x=city_estheticians.values, y=city_estheticians.index, palette="magma")

plt.xlabel('Number of Users')
plt.ylabel('City')
plt.title('City with the Most Users')
plt.show()
```
